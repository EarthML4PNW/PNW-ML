```python
import h5py
import numpy as np
import pandas as pd
```

```python
f = h5py.File("/data/whd01/yiyu_data/PNWML/miniPNW/waveforms.hdf5", "r")
f2 = h5py.File("/data/whd01/yiyu_data/PNWML/mesoPNW/waveforms.hdf5", "w")
```

```python
f2.create_group('data_format')
f2.create_group('data')
f2['data_format']['component_order'] = f['data_format']['component_order'][()]
```

```python
df = pd.read_csv("/data/whd01/yiyu_data/PNWML/miniPNW/metadata.csv")
df["bucket"] = df.apply(lambda r: int(r.trace_name.split("$")[0][6:]), axis = 1)
df["tindex"] = df.apply(lambda r: int(r.trace_name.split("$")[1].split(",")[0]), axis = 1)

df2 = pd.DataFrame(columns=df.columns)
```

```python
for bucket in range(1, 11):
    new_index = df[df["bucket"] == bucket].tindex.max()//10
    print(bucket, new_index)
    df2 = pd.concat([df2, df[(df["bucket"] == bucket) & (df["tindex"] < new_index)]])
    
    f2['data'][f"bucket{bucket}"] = f['data'][f"bucket{bucket}"][:new_index, :, :]
df2.sort_index(inplace=True)
```

```python
df2.to_csv("/data/whd01/yiyu_data/PNWML/mesoPNW/metadata.csv", index=None)
```

```python
f.close()
f2.close()
```
