```python
import copy
# from geopy.distance import geodesic
from obspy.core.utcdatetime import UTCDateTime as utc
from datetime import datetime
import time
import pickle
import h5py
import obspy
import os
import numpy as np
from tqdm import tqdm
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
```

```python
nbucket = 10
n = 0


ix = {}
dataall = {}
dfall = pd.DataFrame()
for year in range(2002, 2022):
    data = {}
    df = pd.read_csv(f"../data/mpiextract/proc{year}_metadata.csv")
    f = h5py.File(f"../data/mpiextract/proc{year}_waveforms.hdf5", 'r')
    d = f['/data/bucket1'][:, :, :]
    f.close()
    
    for idx, i in tqdm(df.iterrows()):
        tn = df.iloc[idx]['trace_name']
        bucket, narray = tn.split('$')
        x, _, _ = iter([int(i) for i in narray.split(',:')])
        array = d[x, :3, :15001]
        
        new_bucket = str(idx % nbucket + 1)
        if new_bucket not in data:
            data[new_bucket] = np.expand_dims(array, axis = 0)
            if year == 2002:
                ix[new_bucket] = 0
        else:
            data[new_bucket] = np.concatenate([data[new_bucket], np.expand_dims(array, axis = 0)], axis = 0)
            ix[new_bucket] += 1
        new_trace_name = f"bucket{new_bucket}${ix[new_bucket]},:3,:15001"
        df.iloc[idx, 18] = new_trace_name
    dfall = dfall.append(df)
    
    for i in range(10):
        if str(i + 1) not in dataall:
            dataall[str(i+1)] = data[str(i+1)]
        else:
            dataall[str(i+1)] = np.concatenate([dataall[str(i+1)], data[str(i+1)]], axis = 0)
    
    data = {}
    df = pd.read_csv(f"../data/mpiextract2/proc{year}_metadata.csv")
    f = h5py.File(f"../data/mpiextract2/proc{year}_waveforms.hdf5", 'r')
    d = f['/data/bucket1'][:, :, :]
    f.close()
    
    for idx, i in tqdm(df.iterrows()):
        tn = df.iloc[idx]['trace_name']
        bucket, narray = tn.split('$')
        x, _, _ = iter([int(i) for i in narray.split(',:')])
        array = d[x, :3, :15001]
        
        new_bucket = str(idx % nbucket + 1)
        if new_bucket not in data:
            data[new_bucket] = np.expand_dims(array, axis = 0)
        else:
            data[new_bucket] = np.concatenate([data[new_bucket], np.expand_dims(array, axis = 0)], axis = 0)
            ix[new_bucket] += 1
        new_trace_name = f"bucket{new_bucket}${ix[new_bucket]},:3,:15001"
        df.iloc[idx, 18] = new_trace_name
        
    dfall = dfall.append(df)
    
    for i in range(10):
        if str(i + 1) not in dataall:
            dataall[str(i+1)] = data[str(i+1)]
        else:
            dataall[str(i+1)] = np.concatenate([dataall[str(i+1)], data[str(i+1)]], axis = 0)
            
```
